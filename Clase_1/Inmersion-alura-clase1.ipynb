## Desafíos

1. Promedio de área de cada uno de los inmuebles en todos los barrios. Gráfico del top 10.
2. Consultar otros datos estadísticos, conteo, mediana, valores mínimos y máximos.

### Tarea 1

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')
```

```python
mean_area_barrios = inmuebles.groupby('Barrio')['Area'].mean() # agrupar por barrio y obtener el promedio del area
mean_area_barrios
```

```python
mean_area_barrios.sort_values(ascending = False, inplace = True) # ordenar valores descendente
top10_barrios = mean_area_barrios.head(10)
top10_barrios
```

```python
# grafico top10
top10_barrios.plot.bar(color = 'purple')
plt.title('Top 10 Barrios')
plt.ylabel('Área promedio m^2')
plt.xlabel('Barrios')
```

```python
(top10_barrios['Modelia']-top10_barrios['Libertador'])*100/top10_barrios['Modelia']
```

Los inmuebles ubicados en el barrio Modelia son los que tienen un mayor área promedio del dataset, siendo de 4316.75 m^2 . Con una diferencia del 85% con respecto a el área promedio del segundo puesto,barrio Libertador.

### Tarea 2: Datos estadísticos

```python
inmuebles['UPZ'].nunique() # cantidad de UPZ en Bogotá
```

```python
inmuebles['Barrio'].nunique() # cantidad de barrios en Bogotá
```

```python
# Número de barrios por UPZ
barrios_upz = inmuebles.groupby('UPZ')['Barrio'].nunique()
barrios_upz
```

```python
# verificando lo anterior
inmuebles[inmuebles['UPZ'] == 'Boyacá Real']['Barrio'].unique()
```

```python
# gráfica número de barrios por UPZ
fig1 = plt.figure()
fig1.set_figwidth(15) # ancho de la figura
barrios_upz.plot.bar(color='purple') # gráfico de barras
plt.title('Número de barrios por UPZ')
plt.xlabel('Unidades de Planeación Zonal')
plt.ylabel('Número de barrios')
```

```python
inmuebles['Tipo'].unique() # qué tipo de inmuebles hay
```

```python
# cantidad de inmuebles por tipo por cada UPZ
tipo_upz_count = inmuebles.groupby('UPZ')['Tipo'].value_counts().unstack(level=-1).fillna(0)
tipo_upz_count
```

```python
tipo_upz_count.plot.bar(figsize=(15, 5),stacked =True) # gráfico de barras apilado
plt.title("Cantidad de inmuebles a la venta las UPZ")
plt.ylabel("Número de inmuebles a la venta")
plt.xlabel("Unidades de Planificación Zonal")
```

Los apartamentos predominan con respecto a los tipos de inmuebles en venta en Bogotá. La UPZ Usaquén tiene la mayor cantidad de inmuebles para la venta.

```python
# Organizar columna de Valor para ser usada
# reemplazar todo lo que no sean números con '' y eliminar los espacios en blanco
inmuebles['Valor'] = inmuebles['Valor'].replace({'[^0-9]':''}, regex = True).str.strip()
inmuebles['Valor'] = inmuebles['Valor'].astype('int64') # cambiar el tipo de dato a entero
inmuebles.dtypes
```

```python
# promedio valor del inmueble por upz
valor_upz = inmuebles.groupby('UPZ')['Valor'].agg(['min', 'max', 'mean', 'median'])
valor_upz
```

En varios casos, el máximo se encuentra muy lejos del promedio; por lo que es mejor tener en cuenta la mediana para el análisis del valor de los inmuebles.

```python
inmuebles.groupby('UPZ')['Valor'].median().plot.bar(figsize=(15, 5))
plt.title("Mediana del Valor del Inmueble")
```

```python
inmuebles[inmuebles['UPZ'] == 'Lucero']['Valor'].median()
```

```python
inmuebles['Valor'].median()
```

```python
inmuebles['Valor'].median()*100/inmuebles[inmuebles['UPZ'] == 'Lucero']['Valor'].median()
```

El valor de los inmuebles ubicados en la UPZ Lucero tienen la mediana más alta en comparación con las demás UPZs. Además, con respecto al dataset, está por encima en un 32%.
